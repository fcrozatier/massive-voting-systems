```python
L1 = [Schulze(500, 15000, CCZip, rematch=3, p=0.9).score() for _ in range(20)]
L2 = [Schulze(500, 15000, CCSlow, rematch=3, p=0.9).score() for _ in range(20)]
```

```python
fig, axes = plt.subplots(1, 2, sharey=True)
fig.suptitle('Schulze benchmark 500 entries, 15k votes')
axes[0].boxplot(L1)
axes[0].set_title('CCZip')
axes[1].boxplot(L2)
axes[1].set_title('CCSlow')


for ax in axes:
  ax.get_xaxis().set_visible(False)
```
