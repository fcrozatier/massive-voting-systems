```python
import matplotlib.pyplot as plt

import os, sys

parent = os.path.abspath("../..")

if not parent in sys.path:
    sys.path.append(parent)

from src.utilities import *
from src.pairings import *
from src.majority_judgment.majority_judgment import *
```

```python
N = 500
budget = 20_000
```

```python
L0 = [MajorityJudgement(N, budget, spread=0.5).score() for _ in range(20)]
L1 = [MajorityJudgement(N, budget, spread=1).score() for _ in range(20)]
L2 = [MajorityJudgement(N, budget, spread=1.5).score() for _ in range(20)]
L3 = [MajorityJudgement(N, budget, spread=2).score() for _ in range(20)]
L4 = [MajorityJudgement(N, budget, spread=2.5).score() for _ in range(20)]
L5 = [MajorityJudgement(N, budget, spread=3).score() for _ in range(20)]
```

```python
fig, axes = plt.subplots(1, 6, sharey=True)
fig.suptitle("Majority Judgement benchmark")

axes[0].boxplot(L0)
axes[0].set_title("sigma=0.5")

axes[1].boxplot(L1)
axes[1].set_title("1")

axes[2].boxplot(L2)
axes[2].set_title("1.5")

axes[3].boxplot(L3)
axes[3].set_title("2")

axes[4].boxplot(L4)
axes[4].set_title("2.5")

axes[5].boxplot(L5)
axes[5].set_title("3")

for ax in axes:
    ax.get_xaxis().set_visible(False)
```
